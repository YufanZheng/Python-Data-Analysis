## Plot Data with Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Plot Series

```python
data = pd.Series(np.random.randn(1000), index=np.arange(1000))
data = data.cumsum()
data.plot()
```

```python
plt.show()
```

### Plot Data Frame

```python
data = pd.DataFrame(np.random.randn(1000,4), index=np.arange(1000), columns=list("ABCD"))
data.head()
```

```python
data = data.cumsum()
data.plot()
plt.show()
```

### Scatter Data Frame

```python
data = pd.DataFrame(np.random.randn(1000, 4), index=np.arange(1000), columns=list("ABCD"))
data = data.cumsum()
```

```python
ax = data.plot.scatter(x='A', y='B', color="Red", label="Class1")
data.plot.scatter(x='A', y='C', color="Green", label="Class1", ax=ax)
plt.show()
```
