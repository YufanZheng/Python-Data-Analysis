# TensorFlow LSTM Regression

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

## Use Numpy to Generate Sequence Data

```python
data = 3 * np.sin(np.linspace(0, 100, 10000))
```

Plot out sequence

```python
plt.figure(figsize=(16, 4))
plt.plot(data)
plt.xticks(())
plt.yticks(())
plt.show()
```

## Prepare Batch Data

```python
num_batch=32

def get_batches(num_batch=num_batch, seq_length=50):
    
    # Use our pre-generated data
    global data
    data_size = data.size
    
    # Input Batches
    x_batches = []
    y_batches = []
    
    for i in range(num_batch):
        # Random find range
        start = np.random.randint(0, data_size-seq_length-1)
        end   = start + seq_length
        # Input is a sequence
        # Target is the next item of the sequence
        x_data = data[start:end]
        y_data = data[end+1]
        # Append data to batches
        x_batches.append(x_data)
        y_batches.append(y_data)
        
    return np.array(x_batches)[:,:,np.newaxis], np.array(y_batches)[:,np.newaxis]
```

---

## Get Weights and Biases

```python
def get_weight(shape):
    init = tf.truncated_normal(shape=shape, stddev=0.1, dtype=tf.float32)
    return tf.Variable(init)

def get_bias(shape):
    init = tf.zeros(shape=shape) + 0.1
    return tf.Variable(init)
```

## Build Network

Hyperparameters

```python
input_size = 1
seq_length = 50
rnn_size   = 64
learning_rate = 0.001
```

Placeholder

```python
with tf.name_scope('Placeholder'):
    xs = tf.placeholder(shape=[None, seq_length, input_size], dtype=tf.float32, name='x_input')
    ys = tf.placeholder(shape=[None, input_size],             dtype=tf.float32, name='y_input')
```

Input Layer

```python
with tf.name_scope('Input_Layer'):
    xs_2d = tf.reshape(xs, [-1, input_size], name='reshape_2_2d')
    Wxs_plus_b = tf.matmul(xs_2d, get_weight([input_size, rnn_size])) + get_bias([rnn_size,])
    xs_in = tf.reshape(Wxs_plus_b, [-1, seq_length, rnn_size], name='reshape_2_3d')
```

RNN Layer

```python
with tf.name_scope('RNN_Layer'):
    lstm_cell = tf.contrib.rnn.BasicLSTMCell(
        num_units=rnn_size, 
        forget_bias=1.0, 
        state_is_tuple=True,
        reuse=tf.AUTO_REUSE
    )
    init_state = lstm_cell.zero_state(batch_size=tf.shape(xs)[0], dtype=tf.float32)
    outputs, final_state = tf.nn.dynamic_rnn(inputs=xs_in, cell=lstm_cell, initial_state=init_state, time_major=False)
    rnn_out = final_state[1]
```

Output Layer

```python
with tf.name_scope('Output_Layer'):
    predictions = tf.layers.dense(inputs=rnn_out, units=1)
```

Loss and train

```python
with tf.name_scope('Loss_and_Train'):
    loss = tf.losses.mean_squared_error(labels=ys, predictions=predictions)
    train = tf.train.AdamOptimizer(learning_rate).minimize(loss)
```

## Train the network

```python
with tf.Session() as sess:
    
    # Init Process
    sess.run(tf.global_variables_initializer())
    sess.run(tf.local_variables_initializer())
    
    for i in range(4001):
        x_input, y_input = get_batches()
        sess.run(train, feed_dict={xs: x_input, ys: y_input})
        if i%100==0:
            print('Loss at step {} : {}'.format(i, sess.run(loss, feed_dict={xs: x_input, ys: y_input})))
            
    # Predict on some value
    for i in range(10):
        index = np.random.randint(0, data.size-seq_length)
        pred_value = sess.run(predictions, feed_dict={xs: data[index:50+index][np.newaxis,:,np.newaxis]})[0][0]
        true_value = data[50+index]
        print('Prediction {} Truth {} Difference {}.'.format(pred_value, true_value, abs(true_value-pred_value)))
```
