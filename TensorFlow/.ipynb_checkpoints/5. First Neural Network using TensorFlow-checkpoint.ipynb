# First Neural Network using TensorFlow

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

## Create dataset

```python
x_data = np.linspace(-1, 1, 300)[:, np.newaxis].astype(np.float32)
noise  = np.random.normal(0.5, 0.05, x_data.shape).astype(np.float32)
y_data = np.square(x_data) - 0.5 + noise.astype(np.float32)
```

Plot the dataset to see the distribution

```python
plt.plot(x_data, y_data)
plt.xlabel("x_data")
plt.ylabel("y_data")
plt.show()
```

## Define a function to add a layer for neural network
- Define weights as a matrix of in_size and out_size
- Define biases as 0.1

```python
def add_layer(input, in_size, out_size, activation_function=None):
    Weights = tf.Variable(tf.random_normal([in_size, out_size]))
    biases = tf.Variable(tf.zeros([1, out_size]) + 0.1)
    Wx_plus_b = tf.matmul(input, Weights) + biases
    if activation_function == None:
        outputs = Wx_plus_b
    else:
        outputs = activation_function( Wx_plus_b )    
    return outputs
```

## Buile the Neural Network

```python
# Placeholder: input and output
xs = tf.placeholder(shape=[None, 1], dtype=tf.float32)
ys = tf.placeholder(shape=[None, 1], dtype=tf.float32)

# Add hidden layer and prediction layer
hidden_layer = add_layer(xs, 1, 10, activation_function=tf.nn.relu)
prediction = add_layer(hidden_layer, 10, 1, activation_function=None)

# Define loss function
loss = tf.reduce_mean( tf.reduce_sum( tf.square(prediction-ys), reduction_indices=[1] ) )

# Train step
optimizer = tf.train.AdamOptimizer(0.01)
train = optimizer.minimize(loss)
```

## Train the network

```python
with tf.Session() as sess:
    
    sess.run(tf.global_variables_initializer())
    
    for step in range(2000):
        sess.run(train, feed_dict={
            xs: x_data,
            ys: y_data
        })
        if step % 200 == 0:
            print(sess.run(loss, feed_dict={
                xs: x_data,
                ys: y_data
            }))
    
    results = sess.run(prediction, feed_dict={ xs: x_data})
```

## Plot out the predict distribution

```python
plt.plot(x_data, results)
plt.xlabel("x_data")
plt.ylabel("y_data")
plt.show()
```
