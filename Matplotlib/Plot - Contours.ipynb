# Matplotlib Plot Contours

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Function to get height

```python
def f(x,y):
    # Get height
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
```

## Prepare Data

```python
n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)

# Mesh Grid
X,Y = np.meshgrid(x,y)
```

## plt.contourf to filling ontours

```python
plt.contourf(X,Y,f(X,Y),8,alpha=0.75, cmap=plt.cm.hot)
```

## plt.contour to add contour lines

```python
C = plt.contour(X,Y,f(X,Y),10, color='black', linewidth=.5)
```

## plt.clabel to add labels

```python
plt.clabel(C,inline=True,fontsize=10)
```

## Plot the Contour line

```python
plt.xticks(())
plt.yticks(())
plt.show()
```
