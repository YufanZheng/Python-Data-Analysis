## Set Legend in Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Prepare the data

```python
x  = np.linspace(-3,2,50)
y1 = x * 2 + 1
y2 = x ** 2
```

```python
plt.figure()

plt.xlim((-2,3))
plt.ylim((-2,5))

plt.xlabel('I am x')
plt.ylabel('I am y')

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top']  .set_color('none')
```

### Plot legend

```python
l1, = plt.plot(x, y1, color='red'  , linewidth=1, linestyle='--', label='up'  ) 
l2, = plt.plot(x, y2, color='green', linewidth=2, linestyle="-" , label='down')

# Default
plt.legend()
# Use self defined
plt.legend(handles=[l1, l2,], labels=['aaa', 'bbb'], loc="upper right")

plt.show()
```
