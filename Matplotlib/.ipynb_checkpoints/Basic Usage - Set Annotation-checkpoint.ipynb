## Matplotlib Set Annotation

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Annotate Data

```python
x = np.linspace(-2, 3, 40)
y = 2 * x + 1

x_point = x[30]
y_point = y[30]

plt.xlim((0,6))
plt.ylim((0,6))

# Plot the line
plt.plot(x, y, color='green', linestyle='-')

# Highlight the point
plt.scatter(x_point, y_point, color='b', s=50)

# Two black dot line
plt.plot([x_point, x_point], [y_point, 0      ], 'k--', lw=2)
plt.plot([0      , x_point], [y_point, y_point], 'k--', lw=2)

# Add annotation
plt.annotate(r'$2x+1=%s$' % y_point, xy=(x_point, y_point), xycoords='data', xytext=(+30, -30),
             textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle="arc3,rad=.2"))

# Add text
plt.text(1.7, 2, r'$This\ is\ the\ some\ text. \mu\ \sigma_i\ \alpha_t$',
         fontdict={'size': 16, 'color': 'r'})

plt.show()
```
