## Matplotlib Figure

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Prepare the data

```python
x = np.linspace(-3, 3, 50)
y1 = x * 2 + 1
y2 = x ** 2
```

### Draw figures

```python
# First figure
plt.figure()
plt.plot(x, y1)
```

```python
# Second figure
plt.figure(num=3, figsize=(8, 5))
plt.plot(x, y2)
plt.plot(x, y1, color="red", linewidth=1.0, linestyle="--")
```

### Show figures

```python
plt.show()
```
