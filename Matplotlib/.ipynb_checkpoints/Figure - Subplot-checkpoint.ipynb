# Matplotlib Figure Subplot 

```python
import matplotlib.pyplot as plt
```

## Create a figure

```python
plt.figure()
```

## Create subplots

```python
plt.subplot(2,2,1)
plt.plot([0,1], [0,1])

plt.subplot(2,2,2)
plt.plot([0,1], [0,2])

plt.subplot(223)
plt.plot([0,1], [0,3])

plt.subplot(224)
plt.plot([0,1], [0,4])

plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot([0,1], [0,1])

plt.subplot(2,3,4)
plt.plot([0,1], [0,2])

plt.subplot(2,3,5)
plt.plot([0,1], [0,3])

plt.subplot(2,3,6)
plt.plot([0,1], [0,4])

plt.show()
```
