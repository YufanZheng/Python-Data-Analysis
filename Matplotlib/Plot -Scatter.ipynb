## Matplotlib Plot Scatter

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Prepare data

```python
# Number of data
N = 1024

# Mean = 0, Stddev = 1
x = np.random.normal(0, 1, N)
y = np.random.normal(0, 1, N)
```

### Scatter Plot

```python
# For color values
T = np.arctan2(y, x)

# c -> color
# s -> size
# alpha
plt.scatter(x,y, c=T, s=15, alpha=0.5)
# plt.scatter(np.arange(5), np.arange(5))

plt.xlim((-1.5, 1.5))
plt.ylim((-1.5, 1.5))

plt.xticks(())
plt.yticks(())

plt.show()
```
