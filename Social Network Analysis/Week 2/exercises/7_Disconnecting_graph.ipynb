# Disconnecting graph

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def plot_graph(G, weight_name=None):
    
    plt.figure()
    pos = nx.spring_layout(G)
    edges = G.edges()
    weights = None
    
    if weight_name:
        weights = [int(G[u][v][weight_name]) for u,v in edges]
        labels = nx.get_edge_attributes(G,weight_name)
        nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
        nx.draw_networkx(G, pos, edges=edges, width=weights);
    else:
        nx.draw_networkx(G, pos, edges=edges)
    
    plt.show()
```

```python
G = nx.karate_club_graph()
plot_graph(G)
```

## Disconnect a graph by remove a node

What is the smallest number of nodes that can be removed from graph to disconnect graph

```python
nx.node_connectivity(G)
```

Which node?

```python
nx.minimum_node_cut(G)
```

See how it looks like after cutting

```python
G = nx.karate_club_graph()
G.remove_node(0)
plot_graph(G)
```

## Disconnect a graph by remove a node

```python
G = nx.karate_club_graph()
```

What is the smallest number of edge to remove for disconnecting

```python
nx.edge_connectivity(G)
```

Which edge?

```python
nx.minimum_edge_cut(G)
```

See the disconnecting graph

```python
G = nx.karate_club_graph()
G.remove_edge(11, 0)
plot_graph(G)
```

## If we remove nodes from the network, could there be no path from 1 to 12?

How many nodes do we need to remove?

```python
nx.node_connectivity(G, 1, 12)
```

What're the nodes?

```python
nx.minimum_node_cut(G, 1, 12)
```

## If we remove edges from the network, could there be no path from 1 to 12?

How many edges do we need to remove?

```python
nx.edge_connectivity(G, 1, 12)
```

What're the edges?

```python
nx.minimum_edge_cut(G, 1, 12)
```
